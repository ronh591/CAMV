```

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

# Define coordinates for the silicon atom and four oxygen atoms
si_coords = [0, 0, 0]
o1_coords = [0.8, 0.8, 0.8]
o2_coords = [-0.8, 0.8, 0.8]
o3_coords = [-0.8, -0.8, 0.8]
o4_coords = [0.8, -0.8, 0.8]

# Create a 3D figure and axes
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the silicon atom as a red sphere
ax.scatter(*si_coords, c='red', marker='o', label='Si')

# Plot the oxygen atoms as green spheres
ax.scatter(*o1_coords, c='green', marker='o', label='O')
ax.scatter(*o2_coords, c='green', marker='o')
ax.scatter(*o3_coords, c='green', marker='o')
ax.scatter(*o4_coords, c='green', marker='o')

# Draw bonds between the silicon and oxygen atoms
ax.plot([si_coords[0], o1_coords[0]], [si_coords[1], o1_coords[1]], [si_coords[2], o1_coords[2]], 'b-')
ax.plot([si_coords[0], o2_coords[0]], [si_coords[1], o2_coords[1]], [si_coords[2], o2_coords[2]], 'b-')
ax.plot([si_coords[0], o3_coords[0]], [si_coords[1], o3_coords[1]], [si_coords[2], o3_coords[2]], 'b-')
ax.plot([si_coords[0], o4_coords[0]], [si_coords[1], o4_coords[1]], [si_coords[2], o4_coords[2]], 'b-')

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Tetrahedral Unit (Si-O)')

# Adjust view angles
ax.view_init(elev=20, azim=-60)

# Show the plot
plt.legend()
plt.tight_layout()
plt.show()
```
